```python
#try7
import dash
from dash import dcc, html, dash_table
from dash.dependencies import Input, Output
import pandas as pd
import numpy as np





max_level = 11  # total number of levels
level_columns = [f"prefLabelLaag{i}" for i in range(1, max_level+1)]

# =============================================================================
# Build the Dash App Layout
# =============================================================================
app = dash.Dash(__name__)

# Generate a list of dropdown containers for levels 1 to 11.
# The first dropdown is always visible; the others are initially hidden.
dropdown_divs = []

# Level 1 dropdown
dropdown_divs.append(
    html.Div([
        html.Label("Selecteer Top Concept:"),
        dcc.Dropdown(
            id="dropdown-1",
            options=[{'label': x, 'value': x} 
                     for x in sorted(filtered_df["prefLabelLaag1"].dropna().unique())],
            placeholder="Selecteer top concept"
        )
    ], style={'width': '30%', 'padding': '10px', 'display': 'block'})
)

# Dropdowns for levels 2 to 11 (initially hidden)
for i in range(2, max_level+1):
    dropdown_divs.append(
        html.Div([
            html.Label(f"Select prefLabelLaag{i}:"),
            dcc.Dropdown(
                id=f"dropdown-{i}",
                placeholder=f"Selecteer concept {i}"
            )
        ], id=f"div-dropdown-{i}", style={'width': '30%', 'padding': '10px', 'display': 'none'})
    )

app.layout = html.Div([
    html.H1("Dashboard Datakwaliteit Cultuurhistorische Thesaurus Concepten"),
    html.P("Selecteer een concept om het onderliggende concept weer te geven."),
    *dropdown_divs,
    html.H2("Samenvatting geselecteerde concept"),
    html.Div(id="summary-div"),
    html.H2("Tabel weergave"),
    dash_table.DataTable(
        id="data-table",
        columns=[{'name': col, 'id': col} for col in merged_df.columns],
        data=merged_df.to_dict("records"),
        page_size=10,
        filter_action="native",
        sort_action="native",
        style_table={'overflowX': 'auto'},
        style_cell={'whiteSpace': 'normal', 'textAlign': 'left'}
    )
], style={'padding': '20px'})

# =============================================================================
# Callback: Update Dropdown Options & Visibility for Levels 2 to 11
# =============================================================================
# This callback uses the current selections from all 11 dropdowns
# (values is a tuple of length 11: [dropdown-1 value, dropdown-2 value, ..., dropdown-11 value])
# For each level i (from 2 to 11), if all previous levels (1 to i-1) have a selection,
# then it filters merged_df accordingly, computes the unique values for level i, and makes its dropdown visible.
# Otherwise, the dropdown is hidden.
from dash.dependencies import Input, Output

# Create a list of inputs for dropdown-1 through dropdown-11.
input_ids = [Input(f"dropdown-{i}", "value") for i in range(1, max_level+1)]

# We need 2 outputs per dropdown for levels 2 to 11:
# one for the options and one for the style (to control visibility).
output_list = []
for i in range(2, max_level+1):
    output_list.append(Output(f"dropdown-{i}", "options"))
    output_list.append(Output(f"div-dropdown-{i}", "style"))

@app.callback(
    output_list,
    input_ids
)
def update_dropdowns(*vals):
    # vals is a tuple with 11 elements: (val1, val2, ..., val11)
    outputs = []
    for i in range(2, max_level+1):
        # For dropdown i, check if all previous levels (1 to i-1) are selected.
        valid = all(vals[j] is not None for j in range(i-1))
        if not valid:
            outputs.append([])  # options: empty list
            outputs.append({'width': '30%', 'padding': '10px', 'display': 'none'})
        else:
            # Filter the DataFrame using selections from levels 1 to i-1.
            df_filtered = merged_df.copy()
            for j in range(1, i):
                df_filtered = df_filtered[df_filtered[f"prefLabelLaag{j}"] == vals[j-1]]
            options = [{'label': x, 'value': x} 
                       for x in sorted(df_filtered[f"prefLabelLaag{i}"].dropna().unique())]
            outputs.append(options)
            outputs.append({'width': '30%', 'padding': '10px', 'display': 'block'})
    return outputs

# =============================================================================
# Callback: Update Summary Table Based on Selections (Deepest Level)
# =============================================================================
@app.callback(
    Output("summary-div", "children"),
    [Input(f"dropdown-{i}", "value") for i in range(1, max_level+1)]
)
def update_summary(*vals):
    # Find the deepest level for which a value is selected.
    deepest = 0
    for i, v in enumerate(vals, start=1):
        if v is not None:
            deepest = i
        else:
            break
    if deepest == 0:
        return "Geen selectie gemaakt."
    df_filtered = filtered_df.copy()
    for i in range(1, deepest+1):
        df_filtered = df_filtered[df_filtered[f"prefLabelLaag{i}"] == vals[i-1]]
    # Use appropriate summary columns:
    if deepest == 1:
        summary_cols = ["prefLabelLaag1", "countWithoutScopeNote"]
    else:
        summary_cols = [f"prefLabelLaag{deepest}",
                        f"mistEigenScopeNote_df{deepest}",
                        f"aantalNarrowerZonderScopeNote_df{deepest}",
                        f"uitklappen_df{deepest}"]
    summary_df = df_filtered[summary_cols].head(1)
    if summary_df.empty:
        return "No details available for this selection."
    return dash_table.DataTable(
        columns=[{'name': col, 'id': col} for col in summary_df.columns],
        data=summary_df.to_dict("records"),
        style_table={"overflowX": "auto"},
        page_size=1
    )

# =============================================================================
# Callback: Update Detailed Data Table Based on Selections
# =============================================================================
@app.callback(
    Output("data-table", "data"),
    [Input(f"dropdown-{i}", "value") for i in range(1, max_level+1)]
)
def update_data_table(*vals):
    df_filtered = merged_df.copy()
    for i, v in enumerate(vals, start=1):
        if v is not None:
            df_filtered = df_filtered[df_filtered[f"prefLabelLaag{i}"] == v]
    return df_filtered.to_dict("records")

# =============================================================================
# Run the App
# =============================================================================
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import dash
from dash import dcc, html, dash_table
from dash.dependencies import Input, Output
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# =============================================================================
# Load Data (assuming filtered_df is already defined)
# =============================================================================
max_level = 11  # Total number of levels
level_columns = [f"prefLabelLaag{i}" for i in range(1, max_level+1)]

# =============================================================================
# Funnel Chart Function
# =============================================================================
def create_funnel_chart(df):
    print("Creating Funnel Chart...")  # Debugging Step
    print(df.head())  # Print first few rows to verify data

    # to numeric
    df["countWithoutScopeNote"] = pd.to_numeric(df["countWithoutScopeNote"], errors="coerce")
    df["aantalNarrowerZonderScopeNote_df2"] = pd.to_numeric(df["aantalNarrowerZonderScopeNote_df2"], errors="coerce")

    # Top Layer
    top_layer = df[["prefLabelLaag1", "countWithoutScopeNote"]].drop_duplicates()
    top_layer.rename(columns={"prefLabelLaag1": "label", "countWithoutScopeNote": "value"}, inplace=True)

    # Second Layer
    second_layer = df[["prefLabelLaag2", "aantalNarrowerZonderScopeNote_df2"]].drop_duplicates()
    second_layer.rename(columns={"prefLabelLaag2": "label", "aantalNarrowerZonderScopeNote_df2": "value"}, inplace=True)

    funnel_data = pd.concat([top_layer, second_layer])


    # Create Funnel Chart
    fig = go.Figure(go.Funnel(
        y=funnel_data["label"],  
        x=funnel_data["value"],  
        textposition="inside",
        textinfo="value+percent initial",
        textfont_size= 20,
        marker=dict(
            color=["#7cc0b9"] * len(funnel_data))
    ))

    fig.update_layout(title="Funnel Chart")

    return dcc.Graph(figure=fig)




# =============================================================================
# Dash App Layout
# =============================================================================
app = dash.Dash(__name__, suppress_callback_exceptions=True)

# list of dropdown levels 1 to 11.
dropdown_divs = []

# Level 1 dropdown
dropdown_divs.append(
    html.Div([
        html.Label("Selecteer Top Concept:"),
        dcc.Dropdown(
            id="dropdown-1",
            options=[{'label': x, 'value': x} 
                     for x in sorted(filtered_df["prefLabelLaag1"].dropna().unique())],
            placeholder="Selecteer top concept",
            style={
        "background-color": "#558f88",  #pp color
        "color": "black",  # Text color
        "border": "1px solid #8FBC8F",  
        "border-radius": "2px",  
        "padding": "2px"
    }
        )
    ], style={'width': '30%', 
              'padding': '10px', 
              'display': 'block'})
)

# Dropdowns 
for i in range(2, max_level+1):
    dropdown_divs.append(
        html.Div([
            html.Label(f"Select prefLabelLaag{i}:"),
            dcc.Dropdown(
                id=f"dropdown-{i}",
                placeholder=f"Selecteer concept {i}"
            )
        ], id=f"div-dropdown-{i}",
        style={'width': '30%', 
                                          'padding': '10px', 
                                          'display': 'none'})
    )

# app layout --------------------------------------------------------------------------------------
app.layout = html.Div([
     dcc.Store(id="stored-data"),

    html.H1("Dashboard Datakwaliteit Cultuurhistorische Thesaurus Concepten"),
    
#------- Funnel Chart -------------------------------------------------------------------------------
    html.Div(
        create_funnel_chart(filtered_df),  
        style={"width": "50%", 
               "float": "right", 
               "padding": "20px", 
               "padding-bottom": "20px", 
               "background-color": "#f5f5f5"}
    ),

    html.P("Selecteer een concept om het onderliggende concept weer te geven."),
    

    dropdown_divs[0],

#-------------------- Dropdowns -----------------------------------------------------
    html.Div([
        html.Div([
            html.Label(f"Select prefLabelLaag{i}:"),
            dcc.Dropdown(
                id=f"dropdown-{i}",
                placeholder=f"Selecteer concept {i}",
                style={ 
                    "background-color": "#7cc0b9",  # pp groen
                    "border": "1px solid #ADDFAD",
                    "border-radius": "2px",
                    "padding": "2px",
                    "color": "black"
                }
            )
        ], id=f"div-dropdown-{i}", style={'width': '30%', 'padding': '10px', 'display': 'none'})
        for i in range(2, max_level+1)
    ]),

    html.H2("Samenvatting geselecteerde concept"),
    html.Div(id="summary-div"),

    html.H2("Tabel weergave"),
    dash_table.DataTable(
        id="data-table",
        columns=[{'name': col, 'id': col} for col in filtered_df.columns],
        data=filtered_df.to_dict("records"),
        page_size=10,
        filter_action="native",
        sort_action="native",
        style_table={'overflowX': 'auto'},
        style_cell={'whiteSpace': 'normal', 'textAlign': 'left'}
        
    ),
        # ✅ Add Download Button
    html.Button("Download CSV", id="download-button", n_clicks=0, style={"margin-top": "10px"}),

    # ✅ Component to Handle CSV Download
    dcc.Download(id="datakwaliteit-cht")
], style={'padding': '20px'})


# =============================================================================
# Callbacks (Dropdowns & Table Updates)
# =============================================================================
input_ids = [Input(f"dropdown-{i}", "value") for i in range(1, max_level+1)]
output_list = []

for i in range(2, max_level+1):
    output_list.append(Output(f"dropdown-{i}", "options"))
    output_list.append(Output(f"div-dropdown-{i}", "style"))

@app.callback(Output("stored-data", "data"), Input("dropdown-1", "value"))
def cache_data(_):
    return filtered_df.to_dict("records")  # data stored in `dcc.Store`

@app.callback(output_list, input_ids)
def update_dropdowns(*vals):
    outputs = []
    for i in range(2, max_level+1):
        valid = all(vals[j] is not None for j in range(i-1))
        if not valid:
            outputs.append([])
            outputs.append({'width': '30%', 'padding': '10px', 'display': 'none'})
        else:
            df_filtered = filtered_df.copy()
            for j in range(1, i):
                df_filtered = df_filtered[df_filtered[f"prefLabelLaag{j}"] == vals[j-1]]
            options = [{'label': x, 'value': x} 
                       for x in sorted(df_filtered[f"prefLabelLaag{i}"].dropna().unique())]
            outputs.append(options)
            outputs.append({'width': '30%', 'padding': '10px', 'display': 'block'})
    return outputs

@app.callback(Output("summary-div", "children"), input_ids)
def update_summary(*vals):
    deepest = 0
    for i, v in enumerate(vals, start=1):
        if v is not None:
            deepest = i
        else:
            break
    if deepest == 0:
        return "Geen selectie gemaakt."
    
    df_filtered = filtered_df.copy()
    for i in range(1, deepest+1):
        df_filtered = df_filtered[df_filtered[f"prefLabelLaag{i}"] == vals[i-1]]

    summary_cols = ["prefLabelLaag1", "countWithoutScopeNote"] if deepest == 1 else [
        f"prefLabelLaag{deepest}", f"mistEigenScopeNote_df{deepest}",
        f"aantalNarrowerZonderScopeNote_df{deepest}", f"uitklappen_df{deepest}"
    ]
    
    summary_df = df_filtered[summary_cols].head(1)
    if summary_df.empty:
        return "Geen concepten gevonden."
    
    return dash_table.DataTable(
        columns=[{'name': col, 'id': col} for col in summary_df.columns],
        data=summary_df.to_dict("records"),
        style_table={"overflowX": "auto"},
        page_size=1
    )

@app.callback(Output("data-table", "data"), input_ids)
def update_data_table(*vals):
    df_filtered = filtered_df.copy()
    for i, v in enumerate(vals, start=1):
        if v is not None:
            df_filtered = df_filtered[df_filtered[f"prefLabelLaag{i}"] == v]
    return df_filtered.to_dict("records")


@app.callback(
    Output("datakwaliteit-cht", "data"),
    Input("download-button", "n_clicks"),
    Input("stored-data", "data"),
    prevent_initial_call=True
)
def download_csv(n_clicks, stored_data):
    # Convert stored data back to a DataFrame
    df = pd.DataFrame(stored_data)
    
    # Convert to CSV format
    return dcc.send_data_frame(df.to_csv, filename="datakwaliteittabel-cht.csv", index=False)

# =============================================================================
# Run the App
# =============================================================================
if __name__ == '__main__':
    app.run_server(debug=False)
```
